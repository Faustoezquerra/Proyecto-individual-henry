Transformaciones solicitadas por el cliente:

1-Desanidando belongs_to_collection

```python

import re

filas_lista = []

for filas in df['belongs_to_collection']:
    if pd.notna(filas) and filas != 'None' and filas != 'Null':
        text = ""
        filas = '['+filas+']'
        diccionarios = ast.literal_eval(filas)
        for diccionario in diccionarios:
            patron = r"'name': '(.*?)'"
            valores_encontrados = re.findall(patron, str(diccionario))
            for valor in valores_encontrados:
                text += valor + ", "
        filas_lista.append(text[:-2])  # Elimina la última coma y espacio
    else:
        filas_lista.append('Sin Dato')

df['belongs_to_collection'] = filas_lista
df['belongs_to_collection'] = df['belongs_to_collection'].replace("","Sin Dato")
df['belongs_to_collection'].head(9)
```

Ahora si desanidando Spoken_Languages para limpiar la columna:

```python
#desanidando spoken languages

import ast

def extract_names(value):
    value = ast.literal_eval(value)  # convierte la cadena en lista
    names = [dic['name'] for dic in value]
    return ', '.join(names)

df['spoken_languages'] = df['spoken_languages'].apply(extract_names)
df['spoken_languages'] = df['spoken_languages'].replace("","Sin Dato")
df['spoken_languages'].head(3)
```

8-Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
# Primero rellenamos los NaN con 0
df['revenue'] = df['revenue'].fillna(0)
df['budget'] = df['budget'].fillna(0)

# Luego convertimos 'revenue' y 'budget' a tipo int
df['revenue'] = df['revenue'].astype(int)
df['budget'] = df['budget'].astype(int)


# Crear la columna 'return' calculándola como 'revenue' / 'budget'
df['return'] = df['revenue'] / df['budget']

# Reemplazar los valores infinitos o NaN con 0
df['return'] = df['return'].replace([np.inf, -np.inf, np.nan], 0)

df.shape
```
